```python
from bs4 import BeautifulSoup
html = """
<html>
<head><title> text와 string의 차이 </title></head>
<p> 
<span>test1</span>
<span>test2</span>
<span><b>test3</b></span>
</p>
</html>
"""
soup = BeautifulSoup(html, "lxml")
tag_p = soup.p # soup 내부의 title 정보를 가져온다. 가장 첫번째 것만 해당됨.
data_text = tag_p.text
data_string = tag_p.string
data_span_str = tag_p.span.string
# text를 이용한 하위 정보 전체 출력
print(data_text)
# string을 이용한 현재 내용에 대해서만 출력
print(data_string)
# string을 이용한 span 태그의 첫번째 줄에 대해서만 출력
print(data_span_str)
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = "https://finance.naver.com/sise/"
page = urlopen(url)
page
```

```python
soup = BeautifulSoup(page, "html.parser")
print(soup)
```

```python
KOSPI = soup.find('span', id="KOSPI_now")
KOSPI.text
```

```python
print("코스피 지수 :", KOSPI.text)
```

```python
# 인기 검색 종목 가격
price = soup.find('ul', class_="lst_pop")
print(price)
price_all = price.find_all('span', class_='up')
#price_all = price.find_all('span')
company_all = price.find_all('a')

price_list = []
for val in price_all:
    #print(val.text)
    price_list.append(val.text)
print(price_list)
```

```python
com_list = []
num_list = []
cnt = 1
for com in company_all:
    #print(com.text)
    num_list.append(cnt)
    com_list.append(com.text)
    cnt = cnt+1
print(num_list)
print(com_list)
```

```python
import pandas as pd
num = pd.Series(num_list)
company_list= pd.Series(com_list)
dat_ori = {"num":num, "company name":company_list}
dat = pd.DataFrame(dat_ori, columns=['num', 'company name'])
dat
```

```python
dat.to_csv("stoc_pop_utf8.csv", index=False, encoding="utf-8") # MAC에서 엑셀파일 생성
dat.to_csv("stoc_pop_euckr.csv", index=False, encoding="EUCKR") #Window에서 엑셀파일 생성
```
