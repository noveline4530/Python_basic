## 일부 정보를 가져오기

```python
from selenium import webdriver

driver2 = webdriver.Chrome('chromedriver')

url = 'https://github.com/TOTOFRIEND/AISTUDY'
driver2.get(url)
```

```python
sel_id = driver2.find_element_by_id("user-content-id01")
print(sel_id)
```

```python
print(sel_id.tag_name)
```

```python
print(sel_id.text)
```

```python
driver3 = webdriver.Chrome('chromedriver')

url = 'https://news.naver.com/'
driver3.get(url)
```

```python
sel_strong = driver3.find_elements_by_tag_name('strong')
len(sel_strong)
```

```python
#text_today_main_news_801001 > li:nth-child(1) > div > a > strong
sel_selector = driver3.find_element_by_css_selector("div.newsnow_tx_inner strong")
print(sel_selector.tag_name)
```

```python
print(sel_selector.text)
```

```python
sel_selector.click()
```

```python
# //*[@id="lnb.searchForm"]/fieldset/input[1]
sel_xpath = driver3.find_element_by_xpath('//*[@id="lnb.searchForm"]/fieldset/input[1]')
```

```python
print(sel_xpath)
```

```python
sel_xpath.clear()
```

```python
sel_xpath.send_keys('미세먼지')
```

```python
sel_xpath = driver3.find_element_by_xpath('//*[@id="lnb.searchForm"]/fieldset/button/span')
sel_xpath.click()

print(driver3.page_source)
```
